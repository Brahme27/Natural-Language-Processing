```python
'''
In text preprocessing, stop words are common words like 
"a," "the," "and," "is," etc., that don’t add much meaning to sentences 
and are often removed to simplify text analysis. 
Removing these words helps focus on more significant words for tasks like search, 
classification, and sentiment analysis.
'''
```

```python
import nltk
```

```python
paragraph="""I have three visions for India. In 3000 years of our history, people from all over the world have come and invaded us, captured our lands, conquered our minds. 
From Alexander onwards, The Greeks, the Turks, the Moguls, the Portuguese, the British, the French, the Dutch, all of them came and looted us, took over what was ours. 
Yet we have not done this to any other nation. We have not conquered anyone. We have not grabbed their land, their culture, their history and Tried to enforce our way of life on them. 
Why? Because we respect the freedom of others.That is why my first vision is that of FREEDOM. I believe that India got its first vision of this in 1857, when we started the war of Independence. 
It is this freedom that we must protect and nurture and build on. If we are not free, no one will respect us.My second vision for India's DEVELOPMENT, For fifty years we have been A developing nation. 
It is time we see ourselves as a developed nation. We are among top 5 nations of the world in terms of GDP. We have 10 percent growth rate in most areas. Our poverty levels are falling. 
Our achievements are being globally recognized today. Yet we lack the self-confidence to see ourselves as a developed nation, self-reliant and self-assured. Isn't this incorrect?
I have a THIRD vision. India must stand up to the world. Because I believe that, unless India stands up to the world, no one will respect us. Only strength respects strength. 
We must be strong not only as a military power but also as an economic power. Both must go hand-in-hand. My good fortune was to have worked with three great minds. Dr. 
Vikram Sarabhai of the Dept. of space, Professor Satish Dhawan, who succeeded him and Dr.Brahm Prakash, father of nuclear material. I was lucky to have worked with all 
three of them closely and consider this the great opportunity of my life.
I see four milestones in my career
"""
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
stopwords.words("english")
```

```python
stopwords.words("german")
```

```python
stopwords.words("arabic")
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
# applying stop words and filter then applying stemming

for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words("english"))]
    # converting all the list of words into sentences with no stop words
    sentences[i]=' '.join(words)
```

```python
sentences
```

```python
# this stemming does not look good 
# lets perform snowballStemming
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowball=SnowballStemmer("english")
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
# applying stop words and filter then applying snowballStemming

for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[snowball.stem(word) for word in words if word not in set(stopwords.words("english"))]
    # converting all the list of words into sentences with no stop words
    sentences[i]=' '.join(words)
```

```python
# it makes all the words in small
sentences
```

```python
# Applying lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
# applying stop words and filter then applying lemmatization

for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words("english"))]
    # converting all the list of words into sentences with no stop words
    sentences[i]=' '.join(words)
```

```python
sentences
```

```python
#Converting into lower case
# applying stop words and filter then applying lemmatization
sentences=nltk.sent_tokenize(paragraph)
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word.lower()) for word in words if word.lower() not in set(stopwords.words("english"))]
    # converting all the list of words into sentences with no stop words
    sentences[i]=' '.join(words)
```

```python
sentences
```

```python
# simple Differences between above three
'''Porter Stemming = basic, fast, sometimes crude stems.
Snowball Stemming = refined, multilingual, slightly more accurate.
Lemmatization = context-aware, produces real words, more accurate but slower.   
'''
```

```python
''' 
Porter Stemming: A basic and widely-used stemming algorithm that chops off common suffixes to reduce words to their stems, 
like turning "running" into "run." However, it can produce stems that aren't actual words (e.g., "consol" from "consolation").

Snowball Stemming: An improved version of Porter Stemming (created by the same developer) 
that's more aggressive and accurate across multiple languages. 
It usually produces cleaner stems than Porter but can still create non-dictionary forms.

Lemmatization: Rather than chopping off parts of words, 
lemmatization reduces words to their root form or lemma, based on the word's meaning and part of speech. 
For example, "better" is lemmatized to "good," while "running" becomes "run." The results are real words, 
making it more accurate but usually slower.

'''
```

```python
lemmatizer.lemmatize("better",pos="a")
```
